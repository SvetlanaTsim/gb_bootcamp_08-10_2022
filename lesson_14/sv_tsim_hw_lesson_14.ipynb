```python
#data analysis libraries 
import numpy as np
import pandas as pd

#visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
#draw a bar plot of sepal_length by species
sns.barplot(x="species", y="sepal_length", data=iris)

#print percentages of females vs. males that survive
#print("Percentage of sepal_length for setosa species:", iris["sepal_length"][iris["species"] == 'setosa'].sum())
print("Percentage of sepal_length for setosa species:", round((iris["sepal_length"][iris["species"] == 'setosa'].sum() / iris["sepal_length"].sum() * 100), 2), "%")
print("Percentage of sepal_length for versicolor species:", round((iris["sepal_length"][iris["species"] == 'versicolor'].sum() / iris["sepal_length"].sum() * 100), 2), "%")            
print("Percentage of sepal_length for virginica species:", round((iris["sepal_length"][iris["species"] == 'virginica'].sum() / iris["sepal_length"].sum() * 100), 2), "%")       
```

```python
#draw a bar plot of petal_length by species
sns.barplot(x="species", y="petal_length", data=iris)

#print percentages of females vs. males that survive
#print("Percentage of sepal_length for setosa species:", iris["sepal_length"][iris["species"] == 'setosa'].sum())
print("Percentage of petal_length for setosa species:", round((iris["petal_length"][iris["species"] == 'setosa'].sum() / iris["petal_length"].sum() * 100), 2), "%")
print("Percentage of petal_length for versicolor species:", round((iris["petal_length"][iris["species"] == 'versicolor'].sum() / iris["petal_length"].sum() * 100), 2), "%")            
print("Percentage of petal_length for virginica species:", round((iris["petal_length"][iris["species"] == 'virginica'].sum() / iris["petal_length"].sum() * 100), 2), "%")   
```

```python
#draw a bar plot of sepal_width by species
sns.barplot(x="species", y="sepal_width", data=iris)

#print percentages of females vs. males that survive
#print("Percentage of sepal_length for setosa species:", iris["sepal_length"][iris["species"] == 'setosa'].sum())
print("Percentage of sepal_width for setosa species:", round((iris["sepal_width"][iris["species"] == 'setosa'].sum() / iris["sepal_width"].sum() * 100), 2), "%")
print("Percentage of sepal_width for versicolor species:", round((iris["sepal_width"][iris["species"] == 'versicolor'].sum() / iris["sepal_width"].sum() * 100), 2), "%")            
print("Percentage of sepal_width for virginica species:", round((iris["sepal_width"][iris["species"] == 'virginica'].sum() / iris["sepal_width"].sum() * 100), 2), "%")    
```
