#### 1. Реализуйте бинарный поиск

```python
def binary_search_recursive(array, element, start, end):
    if start > end:
        return -1

    mid = (start + end) // 2
    if element == array[mid]:
        return mid

    if element < array[mid]:
        return binary_search_recursive(array, element, start, mid-1)
    else:
        return binary_search_recursive(array, element, mid+1, end)
```

```python
array_1 =  sorted([random.randint(0, 100) for i in range(10)])
el = random.sample(array_1, 1)[0]
print(f'Изначальный список {array_1}, нужно найти элемент {el}')
print(f'Найден элемент {binary_search_recursive(array_1, el, 0, (len(array_1)-1))}')
```

#### 2. Реализуйте несколько алгоритмов сортировки и сравните их время работы на одних и тех же входных данных


```python
#сортировка пузырьком (O(n^2))
def bsort(array):
    length = len(array)
    for i in range(length):
        for j in range(length-i-1):
            if array[j] > array[j+1]:
                array[j], array[j+1] = array[j+1], array[j]
    return array
```

```python
#сортировка вставками (O(n^2))
def insertion_sort(nums):
    # Сортировку начинаем со второго элемента, т.к. считается, что первый элемент уже отсортирован
    for i in range(1, len(nums)):
        item_to_insert = nums[i]
        # Сохраняем ссылку на индекс предыдущего элемента
        j = i - 1
        # Элементы отсортированного сегмента перемещаем вперёд, если они больше
        # элемента для вставки
        while j >= 0 and nums[j] > item_to_insert:
            nums[j + 1] = nums[j]
            j -= 1
        # Вставляем элемент
        nums[j + 1] = item_to_insert
    return nums
```

```python
#сортировка выбором (O(n^2))
def selectionsort(arr):
    n = len(arr)
    for i in range(n-1):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr
```

```python
#быстрая сортировка (O(n*log(n)), в худшем случае, когда все элементы надо переставлять (O(n^2)))
def quicksort(nums):
    if len(nums) <= 1:
        return nums
    else:
        q = random.choice(nums)
        s_nums = []
        m_nums = []
        e_nums = []
        for n in nums:
            if n < q:
                s_nums.append(n)
            elif n > q:
                m_nums.append(n)
            else:
                e_nums.append(n)
    return quicksort(s_nums) + e_nums + quicksort(m_nums)
```

```python
#10 элементов
N = 10
```

```python
%%time
#сортировка пузырьком
array_1 =  [random.randint(0, 100) for i in range(N)]
print('Cортировка пузырьком')
print(f'Изначальный список {array_1}')
print(f'Отсортированный список {bsort(array_1)}')
```

```python
%%time
#сортировка вставками
print('Cортировка вставками')
array_1 =  [random.randint(0, 100) for i in range(N)]
print(f'Изначальный список {array_1}')
print(f'Отсортированный список {insertion_sort(array_1)}')
```

```python
%%time
#сортировка выбором
print('Cортировка выбором')
array_1 =  [random.randint(0, 100) for i in range(N)]
print(f'Изначальный список {array_1}')
print(f'Отсортированный список {selectionsort(array_1)}')
```

```python
%%time
#Быстрая сотртировка
print('Быстрая сотртировка')
array_1 =  [random.randint(0, 100) for i in range(N)]
print(f'Изначальный список {array_1}')
print(f'Отсортированный список {quicksort(array_1)}')
```
